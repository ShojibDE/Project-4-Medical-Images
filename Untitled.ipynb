```python
# Use image generator to generate tensor images data and normalize them
# Use 20% of the data for cross-validation  
image_generator = ImageDataGenerator(rescale = 1./255, validation_split= 0.2)
```

```python
# Generate batches of 40 images
# Total number of images is 133*4 = 532 images
# Training is 428 (80%) and validation is 104 (20%)
# Perform shuffling and image resizing
train_generator=image_generator.flow_from_directory(batch_size=40, directory=path, shuffle= True,target_size=(256,256), class_mode = 'categorical', subset="training")
```

```python
# labels Translator 
label_names = {0 : 'Covid-19', 1 : 'Normal' , 2: 'Viral Pneumonia', 3 : 'Bacterial Pneumonia'}
```

```python
# Create a grid of 36 images along with their corresponding labels
L = 6
W = 6

fig, axes = plt.subplots(L, W, figsize = (12, 12))
axes = axes.ravel()

for i in np.arange(0, L*W):
    axes[i].imshow(train_images[i])
    axes[i].set_title(label_names[np.argmax(train_labels[i])])
    axes[i].axis('off')

plt.subplots_adjust(wspace = 0.5)    
```

```python
#Now I use a pre trained model ,so I download it now
basemodel= ResNet50(weights='imagenet', include_top=False, input_tensor=Input(shape=(256,256,3)))
```

```python
basemodel.summary()
```

```python
#freezing the model upto the last stage - 4 and re-training stage -5
for layer in basemodel.layers[:-10]:
    layers.trainable = False
    
```

```python
#Here I buid and train my deep learning model

headmodel1= basemodel.output
headmodel1= AveragePooling2D(pool_size=(4,4))(headmodel1)
headmodel1=Flatten(name='flatten')(headmodel1)
headmodel1=Dense(256, activation='relu')(headmodel1)#I select 256 neuron randomly.
headmodel1=Dropout(0.3)(headmodel1)#DroupOut use for model generalization, here we drop 30% of the neuron. cause Deep learning model is tend to ovefitting.and if i use all the neuron , it literally memorize the all data. so I drop some neuron
headmodel1=Dense(4, activation='softmax')(headmodel1)#we have 4 types image. so I use 4  neuron 

```

There is no fixed things which can improve the model accuracy,so we always have to play with the model architectures. may be we change the number of neuron , or change the Dropout values. etc

```python
#Now I use my pre-trained model and my brand new model
model=Model(inputs= basemodel.input, outputs=headmodel1)
```

```python
# Now I compiling the model
#I have cateroy porblem here , and also more than two category , i mean class so I choose  'categorical_crossentropy'

model.compile(loss = 'categorical_crossentropy', optimizer=optimizers.RMSprop(learning_rate = 1e-4, decay = 1e-6), metrics= ["accuracy"])
```

```python
# using early stopping to exit training if validation loss is not decreasing even after certain epochs (patience)
earlystopping = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=20)
```

```python
# save the best model with lower validation loss
checkpointer = ModelCheckpoint(filepath="weights.hdf5", verbose=1, save_best_only=True)
```

```python
train_generator = image_generator.flow_from_directory(batch_size = 4, directory= path, shuffle= True, target_size=(256,256), class_mode= 'categorical', subset="training")
val_generator = image_generator.flow_from_directory(batch_size = 4, directory= path, shuffle= True, target_size=(256,256), class_mode= 'categorical', subset="validation")
```

```python
history = model.fit_generator(train_generator, steps_per_epoch= train_generator.n // 4, epochs = 10, validation_data= val_generator, validation_steps= val_generator.n // 4, callbacks=[checkpointer, earlystopping])
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

prediction = []
original = []
image = []

for i in range(len(os.listdir(path))):
  for item in os.listdir(os.path.join(path,str(i))):
    img= cv2.imread(os.path.join(path,str(i),item))
    img = cv2.resize(img,(256,256))
    image.append(img)
    img = img / 255
    img = img.reshape(-1,256,256,3)
    predict = model.predict(img)
    predict = np.argmax(predict)
    prediction.append(predict)
    original.append(i)
```

```python
L = 5
W = 5

fig, axes = plt.subplots(L, W, figsize = (12, 12))
axes = axes.ravel()

for i in np.arange(0, L*W):
    axes[i].imshow(image[i])
    axes[i].set_title('Guess={}\nTrue={}'.format(str(label_names[prediction[i]]), str(label_names[original[i]])))
    axes[i].axis('off')

plt.subplots_adjust(wspace = 1.2) 
```

```python
print(classification_report(np.asarray(original), np.asarray(prediction)))
```

```python
cm = confusion_matrix(np.asarray(original), np.asarray(prediction))
ax = plt.subplot()
sns.heatmap(cm, annot = True, ax = ax)

ax.set_xlabel('Predicted')
ax.set_ylabel('Original')
ax.set_title('Confusion_matrix')
```
